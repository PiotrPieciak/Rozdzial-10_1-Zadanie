```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
dataset = pd.read_csv('titanic.csv')
```

```python
fare_data = dataset.loc[dataset['Fare'].notnull(), 'Fare'].values
print(fare_data.shape)
```

```python
plt.hist(fare_data, bins=50)
plt.show()
```

```python
mean_fare = np.mean(fare_data)
print(mean_fare)
```

```python
median_fare = np.median(fare_data)
print(median_fare)
```

```python
mode_fare = stats.mode(fare_data)
print(mode_fare)
```

```python
q0 = np.quantile(fare_data, 0.0)
q1 = np.round(np.quantile(fare_data, 0.25))
q3 = np.round(np.quantile(fare_data, 0.75))
q4 = np.round(np.quantile(fare_data, 1.0))
print(f'Q0: {q0}')
print(f'Q1: {q1}')
print(f'Q2: {median_fare}')
print(f'Q3: {q3}')
print(f'Q4: {q4}')
```

```python
plt.hist(fare_data, bins=50)
plt.axvline(x = mean_fare, color='red', label = 'Średnia')
plt.axvline(x = median_fare, color='green', label = 'Mediana')
plt.axvline(x = mode_fare[0], color='orange', label = 'Moda')
plt.axvline(x = q0, color='black', label = 'Min')
plt.axvline(x = q1, color='black', label = 'Q1')
plt.axvline(x = q3, color='black', label = 'Q3')
plt.axvline(x = q4, color='black', label = 'Max')
plt.legend(loc='upper right')
plt.show()
```

```python
range_fare = max(fare_data) - min(fare_data)
print(range_fare)
```

```python
iqr = q3 - q1
print(iqr)
```

```python
plt.boxplot(fare_data)
plt.show()
```

```python
variance_fare = np.var(fare_data, ddof=1)
print(variance_fare)
```

```python
standard_deviation_fare = np.std(fare_data, ddof=1)
print(standard_deviation_fare)
```
